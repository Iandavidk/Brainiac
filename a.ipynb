```python
import csv
import json

csvfile = open('alpha50.csv', 'r')
k = csv.reader(csvfile)
```

```python
alphas = []
i = 0
for row in k:
    if i == 0:
        i  =1
        continue
    print(row[1])
    alp = json.loads(row[1].replace("'", "\""))
    alp['code'] = row[2]
    alphas.append(alp)
```

```python
alphas
a_new = []
for alpha in alphas:
    a_new.append({
        'code': alpha['code'],
        'region': alpha['Region'],
        'universe': alpha['Universe'],
        'decay': int(alpha['Decay']),
        'truncation': float(alpha['Truncation']),
        'neutralization': alpha['Neutralization'].upper(),
        'pasteurization': alpha['Pasteurization'].upper(),
        'nanHandling': alpha['NaN_Handling'].upper(),
    })
```
